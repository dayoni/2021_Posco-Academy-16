```python
mu = 2

prob = stats.poisson.pmf(3, mu)
cdf_prob = stats.poisson.cdf(2, mu)

print("1분당 {0}번의 전화가 결려 올 확률 : {1:.3f}".format(3, prob))
print("1분당 최대 {0}회 이하의 전화가 결려 올 확률 : {1:.3f}".format(2, cdf_prob))
```

```python
mu = 3

prob = stats.poisson.pmf(2, mu)
cdf_prob = stats.poisson.pmf(1, mu)

print("{0}명의 손님이 도착할 확률 : {1:.3f}".format(3, prob))
print("{0}명 이하의 손님이 도착할 확률 : {1:.3f}".format(2, prob))
print("{0}명 이상의 손님이 도착할 확률 : {1:.3f}".format(2, 1 - cdf_prob))
```

```python
df = pd.DataFrame({'sample':[74.5, 81.2, 73.8, 82.0, 76.3, 75.7, 80.2, 72.6, 77.9, 82.8]})

shapiro(df)
```

```python
t_result = stats.ttest_1samp(df, 76.7)

t, p = t_result.statistic.round(3), t_result.pvalue.round(3)

print("Sample t-test")
print("t 통계량 : {0}".format(t))
print("p-value : {0}".format(p))
```
