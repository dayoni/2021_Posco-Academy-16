## 문제 1  
- #### 파일에 있는 각각의 단어 수 구하기
- #### <text.txt>
    + ##### first line
    + ##### second line
    + ##### third line

```python
f = open('test.txt', 'r')
words = f.read().split()
text_dict = dict()
    
for key in words:
    if key in text_dict.keys() :
        text_dict[key] += 1
    else:
        text_dict[key] = 1

for word, cnt in text_dict.items():
    print(word, cnt)

f.close()
```

## 문제 2  
- #### 파일명을 입력 받아, 해당 파일을 한 줄씩 읽어 파일의 내용을 모두 대문자로 출력하는 프로그램을 작성하시오.
- #### 단, 파일이 없는 경우 “파일이 존재하지 않는다않는다" 정도의 아래 메시지를 출력할 것!

```python
import os

name = input("Enter a file name : ") + ".txt"
if os.path.exists(name) == False:
    print("파일이 존재하지 않습니다.")

else:
    f = open(name, 'r')
    for line in f:
        print(line.upper(), end='')

f.close()
```

## 문제 3  


```python
score = open('score.txt', 'r')
score_list, total, grade, result = list(), list(), list(), list()
text = ""

for line in score:
    score_list.append(line.split())

for i in range(0, len(score_list)):
    for j in range(1, len(score_list[i])-1):
        _total = (float(score_list[i][j])*0.4) + (float(score_list[i][j+1])*0.6)
        total.append(_total)
    if total[i] >= 90.0: grade.append("(A)")
    elif 80.0 <= total[i] < 90.0: grade.append("(B)")
    elif 70.0 <= total[i] < 80.0: grade.append("(C)")
    elif 60.0 <= total[i] < 70.0: grade.append("(D)")
    else: grade.append("(F)")
        
report = open('report.txt', 'w')

for student in range(len(score_list)):
    for cnt in range(len(score_list[student])):
        text = score_list[student][cnt] + " "
        print(text, end=" ")
        report.write(text)
        
    text = str(total[student]) + str(grade[student]) +"\n"
    print(text, end=' ')
    print()
    report.write(text)

score.close()
report.close()
```
