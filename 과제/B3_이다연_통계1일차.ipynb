## 문제 2  
- #### 다음 데이터는 자동차의 모델과 연비(mpg)에 대한 자료를 조사한 Table 이다.
- #### 이 범주형 데이터에 대해 counts, percents, cumulative counts, cumulative percent를 계산 바랍니다.

```python
ds_mycar2 = pd.read_csv('./mycars.csv')
df = ds_mycar2[['model', 'mpg']]

count = df['model'].value_counts().sort_index()

cumcnt = np.cumsum(count)
percent = count / sum(count) * 100
cumpct = np.cumsum(percent)

count_data = pd.DataFrame({'Count':count, 'CumCnt':cumcnt, 'Percent':percent,'CumPct':cumpct})
count_data.columns.name='mpg'
count_data
```

```python
count2 = df['mpg'].value_counts().sort_index()
cumcnt2 = np.cumsum(count2)

percent2 = count2 / sum(count2) * 100
cumpct2 = np.cumsum(percent2)

count_data2 = pd.DataFrame({'Count':count2, 'CumCnt':cumcnt2, 'Percent':percent2, 'CumPct':cumpct2})
count_data2.columns.name='mpg'

count_data2
```

## 문제 3
- #### 어떤 자동차 1리터 당 주행거리가 평균 12km, 표준편자 3km인 정규분포를 이룬다.
1. 1리터를 가지고 9km 이상 15km 이하를 달릴 확률은 얼마 인가요?
2. 1리터를 가지고 15km 이상을 달릴 확률은 얼마 인가요?

```python
mu = 12
sigma = 3
x = 15

prob = stats.norm.cdf(x, mu, sigma)
print("{0}km 이상 {1}km 이하를 달릴 확률은 {2:.3f}에 해당함".format(9, x, ((prob - 0.5) * 2)))
print("{0}km 이상을 달릴 확률 : {1:.3f}에 해당함".format(x, (1-prob)))
```
