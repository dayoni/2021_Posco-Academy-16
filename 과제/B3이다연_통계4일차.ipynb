# <span style="color:red">문제 1</span>
- #### 정규분포에서 추출한 10개의 무작위 데이터로 slice, loc, iloc, query 등을 이용하여 데이터를 선택하라.
- #### 데이터 : 정규분포에서 추출한 10개의 무작위 데이터
- #### 변수 : 4개, 자료 수 : 10개, random seed = 1234
- #### 17page

### 문제풀이
    1. 데이터 구성하기
    2. slice를 이용한 데이터 선택
       - B열 선택, B, D열 선택, 4~8행 선택
    3. loc, iloc를 이용한 데이터 선택
       - loc : 2~8행 A,C열 선택, 3~8행 선택, 3행 선택(단일행)
       - iloc : 0~6행 A열 선택, 2~6행 선택 , 5행 선택(단일행)
    4. query를 이용한 데이터 선택
       - D > 0.5 데이터 선택, B > D 데이터 선택 => slice 표현으로 선택
       - A > 0.5 & D < 0.0 데이터 선택, A > 0.5 | D < 0.0 데이터 선택

```python
# 라이브러리, 모듈 import
import pandas as pd
import numpy as np
from numpy.random import randn
```

```python
# seed(1234)로 하여 랜덤 데이터 생성
np.random.seed(1234)
df = pd.DataFrame(randn(10, 4), columns=['A', 'B', 'C', 'D'])
```

```python
# 데이터 특성 파악
df.describe()
```

```python
# B와 D열의 데이터 출력해보기
print('*'*25)
print(df['B'])
print('*'*25)
print(df['D'])
print('*'*25)
```

```python
# Slice하여 4~8행 선택
print(df[4:9])
```

```python
# loc를 이용하여 데이터 선택
# A의 2~8행 선택
A1 = df.loc[2:8]['A']
print(A1)
```

```python
# C의 2~8행 선택
C1 = df.loc[2:8]['C']
print(C1)
```

```python
# df의 2~6행 선택
df.loc[2:6]
```

```python
# df의 3행 선택(단일행)
df.loc[3:3]
```

```python
# iloc를 이용하여 데이터 선택
# A열 0~6행 선택
A2 = df.iloc[0:7,[0]]
print(A2)
```

```python
# 2~6행 선택
df2 = df.iloc[2:7]
print(df2)
```

```python
# 5행 선택(단일행)
df3 = df.iloc[5:6]
print(df3)
```

```python
# query D > 0.5 데이터 선택을 slice 표현으로 선택
df[df.D < 0.5]
```

```python
# query B > D 데이터 선택을 slice 표현으로 선택
df[df.B > df.D]
```

```python
# query A > 0.5 그리고 D < 0.0 데이터 선택
df.query('A > 0.5 & D < 0.0')
```

```python
# query A > 0.5 혹은 D < 0.0 데이터 선택
df.query('A > 0.5 | D < 0.0')
```

# <span style="color:red">문제 2</span>
- #### 아래와 같은 샘플 데이터 df4, df5, df6을 이용해 concat(좌우 결합)을 실행하라
- #### 데이터 : 직접 생성한 아래와 같은 df4, df5, df6 데이터 (변수 4개, 자료 수 4개)
- #### 32page

### 문제풀이
    1. 데이터 구성하기
    2. concat을 이용한 데이터 결합
    3. parameter 변경

```python
df4 = pd.DataFrame({'A':['A0','A1','A2','A3'],
                    'B':['B0','B1','B2','B3'],
                    'C':['C0','C1','C2','C3'],
                    'D':['D0','D1','D2','D3']})
df4
```

```python
df5 = pd.DataFrame({'E':['E4','E5','E6','E7'],
                    'F':['F4','F5','F6','F7'],
                    'G':['G4','G5','G6','G7'],
                    'H':['H4','H5','H6','H7']})
df5
```

```python
df6 = pd.DataFrame({'I':['I8','I9','I10','I11'],
                    'J':['J8','J9','J10','J11'],
                    'K':['K8','K9','K10','K11'],
                    'L':['L8','L9','L10','L11']})
df6
```

```python
# concat을 이용한 df4, df5, df6 세로(위 아래)결함
result = pd.concat([df4, df5, df6])
result
```

```python
# concat의 parameter 변경 + 가로(좌 우) 결합
result = pd.concat([df4, df5, df6], axis=1, ignore_index=True)
result.columns = ["col_" + str(i) for i in range(12)]
result
```

# <span style="color:red">문제 4</span>
- #### 아래는 운동에 따른 혈당 산소 요구량 데이터이다. pivot을 이용하여 GENDER와 AGEGROUP별 WEIGHT의 평균을 아래와 같이 변환하라.
- #### 데이터1 : FITNESS_1.CSV(변수 5개, 자료 수 31개)
- #### 54page

### 문제풀이
    1. 데이터 구성하기
    2. groupby, agg를 이용한 평균값 계산
    3. pivot을 이용한 변환

```python
# 데이터 구성하기
df1 = pd.read_csv('./FITNESS.csv', encoding='euc-kr')
df1.head()
```

```python
# 'GENDER' 칼럼의 1번 index에 이상치가 들어있으므로 수정한다.
df1.loc[1,'GENDER'] = 'F'
df1
```

```python
# Groupby를 통해 성별과 나이대 별 Weight의 평균을 구한다.
gdf = df1.groupby(['AGEGROUP', 'GENDER'])['WEIGHT'].mean().unstack()
print(gdf.T)
```

```python
# 앞에서 구한 결과를 바탕으로 pivot table을 이용하여 반환하면 다음과 같다.
result = pd.pivot_table(df1, index = 'AGEGROUP', columns = 'GENDER', values = 'WEIGHT', aggfunc = 'mean')
result
```

### 결과
    1. 성별과 나이대 별 평균 weight는 위의 pivot_table과 같다.
